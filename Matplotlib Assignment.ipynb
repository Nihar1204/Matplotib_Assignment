Q1

Matplotlib is a powerful plotting library in Python and it is used to visualize the data. Five plots that can be plotted using Matplotlib: 1 Line Plot: It shows data points connected by straight lines. 2 Scatter Plot: It shows individual data points as markers, useful for visualizing relationship between variables. 3 Bar Plot: Displays categorical data with rectangular bars, often used for comparisons. 4 Histogram: Represents the distribution of a continuous numerical variable by dividing data into bins. 5 Pie Chart: Illustrates data as a circular statistical graphic divided into slices to show numerical proportions. 

```python
# Q2

# Scatter plot is a plotting technique in python. It shows individual data points as markers, useful for visualizing 
# relationship between variables.

import numpy as np
import matplotlib.pyplot as plt
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('X and Y relation')
plt.scatter(x,y)
plt.show()
```

```python
# Q3

# The subplot() function in Matplotlib is used to create a grid of subplots within a single figure. It allows us to arrange
# multiple plots or axes in a grid-like structure, enabling the visualization of several plots within the same figure canvas.

# Data for line 1
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

# Data for line 2
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

# Data for line 3
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

# Data for line 4
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

plt.figure(figsize=(10,8))

# subplot 1
plt.subplot(2,2,1)
plt.plot(x1,y1,marker='o',color='blue')
plt.title('Line 1')

# subplot 2
plt.subplot(2,2,2)
plt.plot(x2,y2,marker='s',color='red')
plt.title('Line 2')

# subplot 3
plt.subplot(2,2,3)
plt.plot(x3,y3,marker='^',color='green')
plt.title('Line 3')

# subplot 4
plt.subplot(2,2,4)
plt.plot(x4,y4,marker='*',color='purple')
plt.title('Line 4')
```

```python
# Q4

# A bar plot is a visualization that represents categorical data with rectangular bars. The length or height of each bar 
# corresponds to the value it represents, making it easy to compare different categories visually.

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Vertical Bar Plot
plt.figure(figsize=(8, 6))
plt.bar(company, profit, color='skyblue')
plt.xlabel('Company')
plt.ylabel('Profit ($)')
plt.title('Company Profits (Vertical Bar Plot)')
plt.show()

# Horizontal Bar Plot
plt.figure(figsize=(8, 6))
plt.barh(company, profit, color='salmon')
plt.xlabel('Profit ($)')
plt.ylabel('Company')
plt.title('Company Profits (Horizontal Bar Plot)')
plt.show()
```

```python
# Q5

# A box plot, also known as a box-and-whisker plot, is a graphical representation of the distribution of a dataset. It 
# displays the five-number summary of a dataset (minimum, first quartile, median, third quartile, maximum), providing insights
# into the data's central tendency, spread, and presence of outliers.

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

data = [box1, box2]


plt.figure(figsize=(8, 6))
plt.boxplot(data)
plt.xticks([1, 2], ['Box 1', 'Box 2'])
plt.xlabel('Box')
plt.ylabel('Values')
plt.title('Box Plot of Box 1 and Box 2')
plt.grid(True)
plt.show()
```
